```python
import numpy as np
import matplotlib.pyplot as plt
from ejercicio1 import plot_vector_field


# ejemplo 1: dy/dx = x + y
def f2(x, y):
    return x + y

# ejemplo 2: dy/dx = sin(x) * cos(y)
def f3(x, y):
    return np.sin(x) * np.cos(y)



print("Ejemplo 1: dy/dx = x + y (Campo unitario)")
fig2 = plot_vector_field(f2, -6, 6, -6, 6, 0.3, 0.3, field_type='N', streamlines=True, density=1.0)
plt.show()

print("Ejemplo 2: dy/dx = sin(x)cos(y) (Campo completo con lneas de flujo)")
fig3 = plot_vector_field(f3, -5, 5, -5, 5, 0.4, 0.4, field_type='F', streamlines=True, density=2.0)
plt.show()
```

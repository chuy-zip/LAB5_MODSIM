```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 400)

# Soluciones con C=0
y1 = x**2 + 1   # y(1) = 2
y2 = x**2 - 3   # y(1) = -2  y(0) = -3
y3 = x**2       # y(1) = 1

plt.figure(figsize=(8,6))
plt.plot(x, y1, label=r"$y(x) = x^2 + 1$, y(1)=2")
plt.plot(x, y2, label=r"$y(x) = x^2 - 3$, y(1)=-2, y(0)=-3", linestyle="--")
plt.plot(x, y3, label=r"$y(x) = x^2$, y(1)=1", linestyle=":")

plt.scatter([1], [2], color="red", zorder=5)    # y(1)=2
plt.scatter([1], [-2], color="blue", zorder=5)  # y(1)=-2
plt.scatter([1], [1], color="green", zorder=5)  # y(1)=1
plt.scatter([0], [-3], color="purple", zorder=5) # y(0)=-3

plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)
plt.title("Soluciones de la EDO con C=0")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.legend()
plt.grid(True)
plt.show()
```
